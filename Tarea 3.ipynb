```python
# Ejercicio 1:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 21.1, 37.8, 54.4, 71.1, 87.8, 100])
y = np.array([0.101, 1.79, 1.13, 0.696, 0.519, 0.338, 0.296])
m = len(x)
n = m - 1

def lagrange_basis(xp, x_points, i):
    L_i = 1
    for j in range(len(x_points)):
        if j != i:
            L_i *= (xp - x_points[j]) / (x_points[i] - x_points[j])
    return L_i

def lagrange_interpolation(xp, x_points, y_points):
    yp = 0
    for i in range(len(x_points)):
        yp += y_points[i] * lagrange_basis(xp, x_points, i)
    return yp

# Solicitar el valor de x para interpolar
p = int(input('Cuatos datos desea interpolar?: '))
xp = []
for i in range(p):
    a = float(input("Enter x: "))
    xp.append(a)
yp = lagrange_interpolation(xp, x, y)

# Crear puntos para la interpolacin
x_interpolado = np.linspace(min(x), max(x), 100)
y_interpolado = [lagrange_interpolation(x_val, x, y) for x_val in x_interpolado]

# Graficar los puntos originales
plt.scatter(x, y, label="Puntos Originales", color="red")

# Graficar el polinomio de interpolacin de Lagrange
plt.plot(x_interpolado, y_interpolado, label="Interpolacin de Lagrange", linestyle="-")

# Graficar los puntos de interpolacin
plt.scatter(xp, yp, label='Puntos interpolados', color="blue", zorder=5)

# Aadir etiquetas y leyenda
plt.xlabel("x")
plt.ylabel("y")
plt.title("Polinomio de Interpolacin de Lagrange")
plt.legend()
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# Ejercicio 2:
from scipy.interpolate import lagrange
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1.525, 3.050, 4.575, 6.10, 7.625, 9.150]
y = [1, 0.8617, 0.7385, 0.6292, 0.5328, 0.4481, 0.3741]

L = lagrange(x, y)

p = int(input('Cuatos datos desea interpolar?: '))
xp = []
for i in range(p):
    a = float(input("Enter x: "))
    xp.append(a)
if max(x) < max(xp):
    x1 = np.linspace(min(x), max(xp), 100) 
else:
    x1 = np.linspace(min(x), max(x), 100)
yp = L(xp)

# graficar valor interpolado
plt.plot(x1, L(x1), label="Interpolacin")
plt.plot(x, y, "o", label="Datos")

# Graficar el valor interpolado
plt.scatter(xp, yp, color="blue",label='Datos interpolados', zorder=5)
plt.legend()
plt.grid()
plt.show()
```
